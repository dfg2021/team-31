```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import wbdata as wb
```

```python
countries = ["NGA","CHN","USA", "GBR", "MEX", "RUS"]
```

```python
indicators = {'EN.ATM.CO2E.KT': 'CO2 emission'}
```

```python
df = wb.get_dataframe(indicators, country='all', data_date=None, freq='Y', source=None, convert_date=False, keep_levels=False, cache=True)
print(df)
```

```python
df.reset_index(inplace = True)
```

```python
df.head(20)
```

```python
df.tail(20)
```

```python
df.describe()
```

```python
dfu = df.unstack(level=0)
```

```python
df_2007=df[df['date']=='2007']
df_2016=df[df['date']=='2016']
```

```python
df_merged = pd.merge(df_2007, df_2016, on=['country'], suffixes=['_2007', '_2016'], how='inner')
```

```python
df_merged.head(6)
```

```python
df_merged.describe()
```

```python
plt.scatter(df_merged['CO2 emission_2007'],df_merged['CO2 emission_2016'])
plt.xlabel('CO2 emission in 2007')
plt.ylabel('CO2 emission in 2016')
```

```python
from bokeh.io import output_file, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import HoverTool
output_notebook()

hover = HoverTool(tooltips=[
     ('country','@country'),
     ])

plot = figure(tools=[hover])
plot.circle('CO2 emission_2007','CO2 emission_2016',
    size=10, source=df_merged)
output_file('inequality.html')
show(plot)
```
